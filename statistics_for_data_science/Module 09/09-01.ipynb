# Time Series

A **time series** is a set of observations at different points in time. A time series can have a **fixed frequency** of observations (e.g. monthly), or have an **irregular frequency** of observations (i.e. the data points are irregularly spaced in time, whenever data is available). In this module, we will only consider time series that are observed at regular intervals of time (e.g. hourly, daily, weekly, monthly, quarterly, annually). Hereafter, we will work with **discretized time** (i.e. we will treat time as a variable, whose value grows incrementally): 

$$t + 1 = t + \Delta t$$


where $\Delta t$ is a **time step** (interval) at which observations of the time series, often referred to as **realizations**, are collected.

Very often, we assume that a time series may involve certain periodicity (e.g. yearly period of flu incidence rates, daily period of temperatures, etc.). In these cases, it is important to specify the time series **frequency** &mdash; i.e. the number of observations within the presumed period. 

For example: 

  - The frequency of daily observations of data with yearly periodicity is $F = 365$
  
  
  - The frequency of hourly observations of data with daily periodicity is $F = 24$

**NOTE:** Proper analysis of time series requires at least two observations per period (i.e. frequency of observations $\geq 2$).

# Techniques of Time Series Analysis

Time series can be subjected to various analyses. Typically, the objective is to obtain a better understanding of the underlying data generating process, or to predict future observations of the time series. The most common techniques used to analyse time series include:

 - Smoothing of the time series
 - Autocorrelation analysis
 - Time series decomposition
 - Modelling the time series and forecasting
 
We will cover these techniques in the following sections.

## Smoothing of the Time Series

**Smoothing** describes a variety of techniques used to reduce, or remove "noise" from a time series. The procedure typically involves calculation of some type of **moving average** (MA) and its subsequent subtraction from the original data, which effectively filters out the short-term fluctuations while preserving the long-term trends.

### Simple moving average 

The most straightforward moving average for smoothing a time series is the **simple moving average** (SMA). The SMA of a signal $y$ at time $t$ is a mean calculated across the set of $k - 1$ consecutive realizations of $y$ preceding time $t$ and the realization at time $t$. Mathematically, this can be formulated as:

$$\text{SMA}_y(t, k) = \frac{y_t + y_{t-1} + \ldots + y_{t- (k-1)}}{k}$$


or, more compactly:

$$\text{SMA}_y(t, k) = \frac{1}{k}\sum^{k-1}_{i=0}\,\, y_{t - i}$$

where:
  
  - $y_t$ is the realization of the time series $y$ at time $t$
  - $\text{SMA}_y(t, k)$ is the SMA of the times series $y$ at time $t$
  - the set of values $y_t, y_{t-1}, \ldots, y_{t- (k-1)}$ comprise the **window** of the moving average
  - $k$ is the window size, sometimes referred to as the **order** of the MA
  
The smoothed signal $y\prime$ is produced from the original time series $y$ by applying the SMA:

$$y_t\prime = \text{SMA}_y(t, k)$$


The size of the window determines "how much" of the signal is to be removed. The bigger the window size, the larger the fluctuations that will be filtered out and the smoother the resulting signal will be.

### Exponential moving average

Sometimes (very often, in fact), to smooth a time series, it is desirable to weight recent realizations of the time series more than older ones. A popular way of doing this is to use an **exponential moving average** (EMA), which for any given time $t$ (assuming $t > 1$), can be calculated by applying the following recursive function:

$$\text{EMA}_y(t, \alpha) = \alpha \cdot y_t + (1 - \alpha) \cdot \text{EMA}_y(t - 1, \alpha)$$

where:

  - $y_t$ and $y_{t-1}$ are the realizations of the time series $y$ at times $t$ and $t-1$ respectively
  - $\text{EMA}_y(t, \alpha)$ is the value of the EMA of $y$ at time $t$
  - $\text{EMA}_y(t - 1, \alpha)$ is the value of the EMA of $y$ at time $t - 1$
  - $\alpha$ is the **exponential smoothing factor**, whose value may range between 0 and 1 ($0 < \alpha < 1$). It represents the degree to which recent realizations are favoured. A greater value of $\alpha$ discounts older realizations faster.
  
**NOTE**: At time $t = 1$, the function needs to be initialized by either setting it equal to $y_1$, or as a mean calculated across the first few realizations of $y$.

Similar to the SMA, the smoothed signal $y\prime$ is produced from the original time series $y$ by applying the EMA:

$$y_t\prime = \text{EMA}_y(t, \alpha)$$

**NOTE**: An important consequence of the recursive character of the EMA is that, depending on the values of $\alpha$ and the initialization method, the early outcomes of the EMA should be disregarded. These outcomes are unreliable, since the EMA requires several iterations to converge to reliable values. This is sometimes called **relaxation** of the EMA.

#### Analogy between EMA and SMA

There is a loose analogy between $k$ of the SMA and $\alpha$ of the EMA, as similar to $k$, bigger values of $\alpha$ will yield a smoother signal. The differences between the original and smoothed signal (sometimes referred to as errors) of $\text{EMA}(t, \alpha)$ and $\text{SMA}(t, k)$ have roughly the same distribution when $\alpha = 2 / (k + 1)$.

Importantly, the EMA takes into account all past data, whereas the SMA only takes into account $k$ past data points. Another key difference between the EMA and the SMA is that calculation of the SMA requires the past $k$ signal realizations, whereas calculation of EMA only requires the most recent realization and the preceding value of the EMA, which makes the EMA less memory demanding and computationally "cheaper."

There are numerous modifications of the SMA and the EMA as well as additional alternatives. For more details about these we recommend visiting: https://en.wikipedia.org/wiki/Moving_average (Moving Average, n.d.).

## Autocorrelation Analysis

**Autocorrelation**, sometimes referred to as serial correlation, is a correlation between the actual values of the time series and its preceding values, as a function of the delay between them. Mathematically, this is expressed as:

$$R_{yy}(\tau) = \frac{E\big[(y_t - \mu)(y_{t + \tau} - \mu)\big]}{\sigma^2}$$


where:

  - $y$ is the time series, sometimes called the **signal**
  - $R_{yy}$ denotes autocorrelation of the time series $y$
  - $\tau$ is a delay between the actual values and preceding ones; and is often called **lag**  
  - $y_t$, $y_{t + \tau}$ are realizations of the time series $y$ recorded at times $t$ and $t + \tau$, respectively
  - $E$ is the **expected value** operator. It is the *de facto* long-term average of the argument: $(y_t - \mu)(y_{t - \tau} - \mu)$
  - $\mu$ and $\sigma$ are the mean and standard deviation of $y$
  
By removing the normalization term $\sigma^2$, we obtain **autocovariance**, which is frequently used as an alternative to autocorrelation:

$$C(\tau) = E[(y_t - \mu)(y_{t - \tau} - \mu)]$$


**NOTE**: In some literature, the term autocorrelation and autocovariance are used interchangeably. However, there are at least two substantial differences between them:

- Autocovariance is resistant to cases when $\sigma = 0$, which makes it useful in cases when we don't know $\sigma$ prior to the calculation of autocorrelation (i.e. when there is a chance that the time series is constant).


- Autocorrelation, in contrast to autocovariance, is normalized &mdash; i.e. it can be used for comparison of time series with very different ranges of values (e.g. a record of two different stocks)
  
Autocorrelation is a special case of a more general function called **cross-correlation**, $R_{yy}(\tau)$:

$$R_{xy}(\tau) = \frac{E\big[(x_t - \mu_x)(y_{t + \tau} - \mu_y)\big]}{\sigma_x \sigma_y}$$

### Properties of autocorrelation

There are several interesting properties of autocorrelation that are worth mentioning:

  - $R_{yy}(\tau)$ is an even function, which means that $R_{yy}(\tau) = R_{yy}(-\tau)$
  - The autocorrelation of a continuous signal reaches its peak at $\tau = 0$, and for any value of $\tau$: $|R_{yy}(\tau)| < R_{yy}(0)$
  - The autocorrelation of a periodic signal is always periodic with the same period
  - The autocorrelation of the sum of two completely uncorrelated signals (i.e. the cross-correlation is equal to zero for all values of $\tau$) equals the sum of the autocorrelations of each signal separately
  - If the autocorrelation of a signal is equal to zero for all $\tau > 0$, we say the signal is **white noise**

### How to estimate autocorrelation

When time is discretized and the values of $\mu$ and $\sigma$ can be calculated beforehand, the autocorrelation can be estimated by the following formula:

$$\hat{R}_{yy}(k) = \frac{1}{(n - k)\sigma^2} \sum_{t = 1}^{n - k}(y_t - \mu)(y_{t + k} - \mu)$$


where:

- $k$ is the lag &mdash; i. e. the number of time intervals between the actual and preceding realizations of $y$ (discrete analogy to $\tau$)
- $n$ is the number of realizations in the given sample &mdash; i.e. the number of data points over which we are going to calculate the autocorrelation
- $\mu$ and $\sigma$ are the mean and standard deviation of $y$ calculated across the sample

### Partial autocorrelation

**Partial autocorrelation** ($\Phi$) measures the autocorrelation of one variable at a given lag, after removing the effect of all shorter lags. 

For example, the partial autocorrelation of lag $\tau = 2$ measures correlation between $y_t$ and $y_{t-2}$ after removing the effect of $y_{t-1}$ on both $y_t$ and $y_{t-1}$. Partial autocorrelation can be defined in multiple ways. Probably the simplest formulation is the following recursive scheme:

$$\Phi_{t}(\tau) = \frac{R_{t}(\tau) - \sum^{\tau - 1}_{j = 1}\,\Phi_{t-1}(j)\cdot R_{t}(\tau - 1)}{1 - \sum^{\tau - 1}_{j = 1}\,\Phi_{t-1}(j)\cdot R_{t}(\tau - 1)}$$


where:

  - $R_t(\tau)$ is the autocorrelation, as measured at time $t$ with lag $\tau$
  - $\Phi_t(\tau)$ is the partial autocorrelation, as measured at time $t$ with lag $\tau$

### Correlogram

Plotting the dependence of $R_{yy}(\tau)$ on $\tau$, we obtain a **correlogram**, or **lag plot**. A correlogram is an extremely useful tool for visual examination of the properties of a time series. Correlograms can be used to assess various important properties of time series, including:

  - Is the observed signal **random**?
  
  
  - Is the observed signal **periodical**?
  
  
  - Is the observed signal **autoregressive**? (can it be modelled by autoregression, covered in a later section)

## Time Series Decomposition

Time series decomposition is a technique that deconstructs the observed time series into several time series components, each of which has certain characteristic properties. Time series are usually decomposed into three components:

1. **Trend** ($T(t)$) reflects the long-term progression of the time series. A trend component is present if there is a persistent increase or decrease of the signal.<br><br>
  
2. **Seasonality** or the seasonal component ($S(t)$) reflects the repeating changes of the time series. A seasonal component is present if the signal changes in repeating cycles with a fixed period. This occurs when the time series is affected by some periodic factors (e.g. seasonality throughout the year, day and night periodicity, etc.).<br><br>
  
3. **Noise** or the irregular component ($I(t)$) reflects the irregularities of the time series. Noise represents the residuals of the observed time series after the other two components have been extracted.
<br>
<blockquote><b>NOTE:</b> In some cases decomposition of the time series involves a fourth component, often referred to as <b>cyclicality</b> or the <b>cyclic</b> component ($C(t)$), that reflects the repeating changes of the times series that, in contrast to those reflected by seasonality, don't have a fixed period. Here we will limit ourselves to cases where the time series is decomposed into just the three above mentioned components, ignoring the cyclic component.</blockquote>

A time series is considered to be a combination of these three components. This combination can be either **additive** or **multiplicative**. Additive decomposition (also known as an **additive model**) assumes the observed time series is a sum of the individual components:

$$y_t = T(t) + S(t) + I(t)$$


Multiplicative decomposition (also known as a **multiplicative model**) assumes the observed time series is a product of the individual components:

$$y_t = T(t) \cdot S(t) \cdot I(t)$$


The additive decomposition is conducted as follows:

1. If frequency of the time series, $F$, is an even number, calculate the estimate of the trend component $\hat{T}(t)$, by calculating the $2 \cdot F$ moving average. If $F$ is an odd number, estimate the trend by calculating the $F$ moving average.<br><br>
  
2. Calculate the **detrended** time series by subtracting the trend estimate: $y_t - \hat{T}(t)$.<br><br>
  
3. Average the detrended values across the presumed period &mdash; i.e. calculate the average across the sets of $F$ consecutive values of $y_t - \hat{T}(t)$ to obtain the estimate of seasonality $\hat{S}(t)$.<br><br>
  
4. Calculate the noise component by subtracting the trend and seasonality estimates: $\hat{I}(t) = y_t - \hat{T}(t) - \hat{S}(t)$.

**NOTE**: Multiplicative decomposition is conducted by an analogous procedure, but division is applied instead of subtraction.

### When to use additive and multiplicative decomposition
    
In general, the additive model is more appropriate if the magnitude of the seasonal component does not vary much with the actual signal. When there is an association between the two, the multiplicative model is more appropriate. Additive decomposition is more common when working with agricultural, epidemiological, or other biological time series. Multiplicative decompositions are more common with financial and economic time series.

### Seasonality adjusted time series

Sometimes, the variation of the time series due to seasonality is not of interest. For example: 

  - Monthly unemployment rates are usually affected by seasonal variation (e.g. an increase in unemployment due to people leaving school and seeking work), thus complicating the ability to assess the underlying state of the labour market
  
  
  - Flu rates are heavily affected by seasonality (e.g. high incidence rate during cold months), which complicates assessment of the underlying epidemiologic situation

Therefore, time series are often adjusted by removing the seasonal component from the original time series, producing the **seasonally adjusted signal**. For an additive decomposition, the seasonally adjusted signal is given by $y_t - S(t)$, and for a multiplicative model, the seasonally adjusted signal is given by $y_t / S(t)$.

### Measuring the strength of trend and seasonality

Sometimes we want to know how "strong" the effect of trend and/or seasonality is in our time series. The strength of a trend can be quantified by assessing its contribution to the overall variance of the signal. For a strongly trended signal, the seasonally adjusted data should have more variation than the remainder component. For an additive model, this can be mathematically formulated as follows:

$$A_T = \text{max}\Big(0, 1 - \frac{\text{Var}\big(I(t)\big)}{\text{Var}\big(T(t) + I(t)\big)}\Big)$$


Analogously, strength of seasonality can be formulated as:

$$A_S = \text{max}\Big(0, 1 - \frac{\text{Var}\big(I(t)\big)}{\text{Var}\big(S(t) + I(t)\big)}\Big)$$


where:

  - $A_T$ and $A_S$  are the strength of trend and seasonality, respectively
  - $T(t)$ is trend
  - $S(t)$ is seasonality
  - $I(t)$ is noise (irregularity component)

For multiplicative models, the denominators in the above equations contain multiplication instead of addition.

## Modelling of Time Series and Forecasting

Naturally, people are often interested in how a time series will continue to evolve in the future. There are numerous methods that have been developed to model the behaviour of times series to predict their future. Most of these methods utilize either regression analysis, some type of moving average, or a combination of both. In the following section, we will discuss the most fundamental of these methods.

### Autoregression models

Time series regression is the application of regression analysis on time series data. The most common type of time series regression is linear regression, where the response is the actual value of the time series, and predictors (features) are its preceding values. This type regression is called **autoregression**, as it uses the historical values of the time series to estimate the values of the same time series.

Autoregression can be mathematically expressed as follows:

$$\hat{y}_t = \beta_0 + \beta_1 y_{t-1} + \beta_2 y_{t-2} + \ldots + \beta_p y_{t-p} + e_t$$


or, more compactly:


$$\hat{y}_t = \beta_0 + \sum^{p}_{i = 1}\beta_i y_{t-i} + e_t$$

where:

  - $\hat{y}$ is the **forecasted value** (or simply the forecast) of the signal $y$ for time $t$
  - $\beta_0$ is the **autoregression constant**. It is the equivalent of the intercept term in previously described regression analysis.
  - $\beta_i$ for $i = 1, 2,\ldots, p$ are the **autoregression coefficients**
  - $p$ is the **order of the autoregression**, indicating the **maximum lag** of the autoregression
  - $e_t$ is called **white noise**. It is the **error term** equivalent to those of regression analysis. White noise can be thought of as the time series whose values are normally distributed with mean equal to zero.

Estimates of the autoregression coefficients constitute an **autoregression model (AR)** of the $p$-th order, often denoted as $\text{AR}(p)$. The fact that the autoregression model is used to predict the future values is then expressed as: $\hat{y}_t = \text{AR}(p)$.

#### The choice of the maximum lag

An important aspect of autoregression is **the choice of the maximum lag** $p$ (order of the autoregression) &mdash; i.e. how far into the past the autoregression goes to predict the future. The maximum lag can be estimated using the **partial autocorrelation** of the times series. 

Typically, as the lag increases, the absolute value of the partial autocorrelation quickly decreases, approaching zero. The lag beyond which all the partial autocorrelations are equal (or very close) to zero can be used as the maximum lag of the autoregression/moving average. However, generally it is safer to use lag that is incremented by one time step from this. Mathematically:

$$p = k + 1,\,|\Phi(i)| = 0\,\,; \forall i > k$$

### Moving average model

The **moving average (MA) model** is another common approach for time series forecasting which, rather than using the past values of the time series, the past forecast errors are used in a regression-like model:

$$\hat{y}_t = \mu_y + \epsilon_{t} + \theta_1 \epsilon_{t-1} + \ldots + \theta_q \epsilon_{t-q}$$


or, more compactly:

$$\hat{y}_t = \mu_y + e_t + \sum^{q}_{i = 1} \theta_i \epsilon_{t-i}$$


where:

- $\hat{y}$ is the **forecasted value** of the signal $y$ for time $t$
- $\mu_y$ is the mean of the time series $y$
- $\theta_i$ for $i = 1, 2,\ldots, q$ are the **model parameters**
- $q$ is the **order of the moving average**
- $\epsilon_{t-i}$  for $i = 1, 2,\ldots, q$ are the white noise **error terms** obtained from the previous forecasts, sometimes called **random shocks**. They are typically assumed to come from the normal distribution with zero mean.
- $e_t$ is the error term (white noise) of the actual prediction, the equivalent of $e_t$ in autoregression
<br>
<blockquote>
<b>NOTE:</b> Error terms of the moving average model are different from those of the autoregression model. Unlike error terms of autoregression, they are affected by the preceding predictions. Hence different notation. 
</blockquote>

The fact that the moving average model of the $q$-th order is used to predict the future values is expressed as: $\hat{y}_t = \text{MA}(q)$

#### A few important notes about the moving average model

- The MA model assumes the time series is **stationary** &mdash; i.e. its mean and standard deviation don't change very much over time. This is a very strong assumption that typically holds only temporarily. Therefore, MA models have only a limited "life span" and after some time, need to be re-estimated.


- Conceptually, the moving average model is a linear regression, but, because we do not observe the $\epsilon$ values directly, it is not really a linear regression in the usual sense. 


- Error terms from the previous time step **propagate** to future predictions, as they become predictors in the next time step. Hence, each error term affects only the next $q$ consecutive estimates of $y$. This gives MA models substantial advantage over AR models, as even major irregularities (spikes, dropdowns) in the time series affect the predictions only in the next few steps.


- The name "moving average model" comes from the fact that each value of $\hat{y}$ can be thought of as a weighted moving average of the past $q$ forecast errors. Importantly, a moving average model must not be confused with moving average smoothing:<br><br>

  - A **moving average model** is used for forecasting future values of a time series<br><br>
  
  - **Moving average smoothing** is used for noise filtering and for trend estimation<br><br>
  
- Fitting an MA model is more complicated than AR model fitting. This is because the lagged error terms $\epsilon$ are not directly observable. Therefore, **iterative methods for non-linear fitting** are applied instead of traditional linear least squares. A description of these methods is not the subject of this material, but can be found here: https://www.it.uu.se/research/publications/reports/2006-022/2006-022-nc.pdf (Sandgren & Stoica, 2006).

#### The choice of the maximum lag

**Order of the moving average model** (maximum lag) can be estimated from the autocorrelation function (not necessarily partial autocorrelation). Similar to the order of the autoregression model, the maximum lag should be equal to the lag beyond which autocorrelation is equal to zero, typically increased by one.

### Autoregressive moving average (ARMA) model (and its extensions)

The **ARMA model** (sometimes called the Box-Jenkins model), is combination of the autoregression and moving average models:

$$\hat{y}_t = \beta_0 + \sum^{p}_{i = 1}\beta_i y_{t-i} + \sum^{q}_{i = 1} \theta_i \epsilon_{t-i} + e_t$$


where the notation remains the same as in the case of autoregression and autocorrelation. The application of the ARMA model to predict $y_t$ is expressed as:

$$\hat{y}_t = \text{ARMA}(p, q)$$


**Importantly**, the ARMA model, similar to MA models, assumes that the time series is stationary &mdash; i.e. its mean and standard deviation don't change very much over time. This assumption typically holds only for a short period of time, after which the mean "drifts away." To overcome this limitation, a generalization of the ARMA model, called **ARIMA** has been proposed.

The **ARIMA model** (Autoregressive integrated moving average model) is a generalization of the ARMA model, that accounts for the non-stationarity of the time series. The ARIMA model is formally identical to ARMA, but requires an initial **differencing step** to be applied (one or multiple times) to the time series $y$ in order to eliminate non-stationarity. **"Differencing of the time series"** means the difference between consecutive observations is computed. Mathematically, this is expressed as:

$$y\prime_t = y_t - y_{t-1}$$


Sometimes second order differencing is applied:

$$y\prime\prime_t = y\prime_t - y\prime_{t-1}$$


Application of the ARIMA model to predict $y_t$ is expressed as:

$$\hat{y}_t = \text{ARIMA}(p, d, q)$$


where $d$ indicates the order of the differencing used prior to the application of the ARMA model. Therefore ARMA, AR, MA can be thought of as special cases of ARIMA, particularly:

  - $\text{ARIMA}(p, 0, q)$ model is $\text{ARMA}(p, q)$
  - $\text{ARIMA}(p, 0, 0)$ model is $\text{AR}(p)$
  - $\text{ARIMA}(0, 0, q)$ model is $\text{MA}(q)$
  - $\text{ARIMA}(0, 0, 0)$ model is **white noise**
  - $\text{ARIMA}(0, 1, 0)$ model ($y_t = y_{t-1} + \epsilon_t$) is a random signal, called a **random walk**
  
There are multiple extensions to the ARMA/ARIMA models, such as SARIMA, ARIMAX, etc., which are not in the scope of this material. More information about these can be found here: https://arxiv.org/pdf/1302.6613.pdf (Adhikari & Agrawal, 2013).

# Long-term Predictions

Sometimes, we need to forecast more than just one time step ahead. This is can be addressed in two ways: 

- **A direct forecast**. This forecast model is designed in a way that it directly estimates the value of $y$ in a more distant future. For example, in the case of an autoregression model it can be formulated as:
  
$$\hat{y}_{t+h} = \beta_0 + \sum^{p}_{i = 0}\,\beta_i\,y_{t} + e_{t+h}$$


where $h$ is called the **prediction horizon**.
    
- **An iterative (inductive) forecast**. This means we will perform a sequence of short-term predictions (one step ahead), where each prediction uses previous prediction estimates as if it were an actual value.

## Word of caution with long-term predictions

Regardless of whether iterative or direct forecast is used, the **predictability** of future values of any time series decreases rapidly with increases of the prediction horizon. How far into the future we can predict the time series values depends on many factors, most prominently:

  - Amount and frequency of observations
  - Strength of seasonality, trend and noisiness
  - Forecasting method
  - Domain specific factors

There are multiple methods for estimating the maximum prediction horizon. For more details see: https://robjhyndman.com/papers/ijf25.pdf (De Gooijer & Hyndman, 2006).

However, **long-term predictions should be avoided unless absolutely necessary**.
