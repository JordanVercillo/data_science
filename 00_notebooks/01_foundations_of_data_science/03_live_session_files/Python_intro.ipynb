# Function

```python
# this function extracts all the int values from any input dictionary
def extract(input_dictionary):
  result_list = []
  for key, val in input_dictionary.items():
    if type(val) == int:
      result_list.append(val)
  return result_list
```

```python
extract({"a": 111, "b":2222, "c": "str"})
```

```python
result = extract({"a": 111, "b":2222, "c": "str"})
result
```

```python
# get the key of the first element in a dictionary whose value is an integer
def find(input_dictionary):
  for key, val in input_dictionary.items():
    if type(val) == int:
      return key  # return the result immediately and stops function

find({"x": "stringgggg", "y": -2, "a": 111, "b":2222, "c": "str"})
```
